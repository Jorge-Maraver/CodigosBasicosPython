```python
#Abstracción
"""
Crear una clase abstracta envase de líquido, que contenga los atributos: capacidad máxima y contenido actual
Debe contener el siguiente método abstracto:
- Rellenar: en las clases hijas, recibe un valor y debe aumentar dicha cantidad el contenido actual sin sobrepasar la capacidad máxima

De éste debe heredar dos clases hijas: vaso y botella
La clase botella debe contener el siguiente métodos:
- Servir: recibe un valor y debe reducir dicha cantidad el contenido actual hasta llegar a 0
"""

#Importación de los elementos necesarios de la librería abc
from abc import ABC,abstractmethod


#Clase abstracta
class EnvaseLiquido(ABC):
    
    #Constructor abstacto
    @abstractmethod
    def __init__(self, capacidad, contenido):
        pass
    
    #Método rellenar abstacto
    @abstractmethod
    def Rellenar(self,cantidad):
        pass
        
class Vaso(EnvaseLiquido):
    
    #Constructor (contenido por defecto a 0)
    def __init__(self, capacidad, contenido=0):
        self.capacidad=capacidad
        self.contenido=contenido
        print("Vaso creado\n")
        
    #Función rellenar
    def Rellenar(self,cantidad):
        self.contenido+=cantidad
        if(self.contenido>self.capacidad):
            self.contenido=self.capacidad
        print(f"El vaso contiene ahora {self.contenido} cl\n")
        
class Botella(EnvaseLiquido):
    
    #Constructor (contenido por defecto a 0)
    def __init__(self, capacidad, contenido=0):
        self.capacidad=capacidad
        self.contenido=contenido
        print("Botella creada\n")
        
    #Función rellenar
    def Rellenar(self,cantidad):
        self.contenido+=cantidad
        if(self.contenido>self.capacidad):
            self.contenido=self.capacidad
        print(f"La botella contiene ahora {self.contenido} cl\n")
        
    #Función servir
    def Servir(self,cantidad):
        if(self.contenido<cantidad):
            #Variable para saber cuánto se ha servido
            servido=self.contenido
            self.contenido=0
        else:
            self.contenido-=cantidad
            #Variable para saber cuánto se ha servido
            servido=cantidad
        print(f"Se han servido {servido} cl.\nLa botella contiene ahora {self.contenido} cl\n")


#Creación de un vaso
v1=Vaso(50)

#Creación de una botella
b1=Botella(50)

#Rellenar el vaso
v1.Rellenar(30)
print("#\n",b1.contenido,"\n#")

#Rellenar la botella
b1.Rellenar(30)
print("#\n",b1.contenido,"\n#")

#Servir con la botella
b1.Servir(20)
print("#\n",b1.contenido,"\n#")

#Creación de un envase líquido (debería dar error porque es de clase abstracta)
try:
    #Creación de un envase "genérico"
    el1=EnvaseLiquido(50,25)
except:
    print("ERROR: No se ha podido completar la instrucción porque la clase envase líquido es absracta.")
```
