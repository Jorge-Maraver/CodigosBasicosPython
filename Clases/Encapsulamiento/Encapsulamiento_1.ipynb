```python
#Encapsulamiento
"""
Crear una clase botella, que contenga los atributos: capacidad máxima y contenido actual
Además, debe contener los siguientes métodos:
- Rellenar: recibe un valor y debe aumentar dicha cantidad el contenido actual sin sobrepasar la capacidad máxima 
- Servir: recibe un valor y debe reducir dicha cantidad el contenido actual hasta llegar a 0

Todo ello con funciones getter y setter
"""

class Botella:
    
    #Constructor (contenido por defecto a 0)
    def __init__(self, capacidad, contenido=0):
        self._capacidad=capacidad
        self._contenido=contenido
        print("Botella creada\n")
    
    #Funciones getter y setter
    @property
    def capacidad(self):
        return self._capacidad
    
    @property
    def contenido(self):
        return self._contenido    
    
    @capacidad.setter
    def capacidad(self,capacidad):
        self._capacidad=capacidad
    
    @contenido.setter
    def contenido(self,contenido):
        self._contenido=contenido
        
    #Función rellenar
    def Rellenar(self,cantidad):
        self.contenido+=cantidad
        if(self.contenido>self.capacidad):
            self.contenido=self.capacidad
        print(f"La botella contiene ahora {self.contenido} cl\n")
        
    #Función servir
    def Servir(self,cantidad):
        if(self.contenido<cantidad):
            #Variable para saber cuánto se ha servido
            servido=self.contenido
            self.contenido=0
        else:
            self.contenido-=cantidad
            #Variable para saber cuánto se ha servido
            servido=cantidad
        print(f"Se han servido {servido} cl.\nLa botella contiene ahora {self.contenido} cl\n")
        
        
#Creación de una botella con el valor por defecto
b1=Botella(50)

#Validación de la función colocar
b1.Rellenar(30)
print("#\n",b1.contenido,"\n#")

#Validación de la función servir
b1.Servir(20)
print("#\n",b1.contenido,"\n#")

#Validación de la función rellenar con demasiada cantidad
b1.Rellenar(100)
print("#\n",b1.contenido,"\n#")

#Validación de la función servir con demasiada cantidad
b1.Servir(100)
print("#\n",b1.contenido,"\n#")

#Creación de una botella sin usar el valor por defecto
b2=Botella(50,30)
print("#\n",b2.contenido,"\n#")
```
