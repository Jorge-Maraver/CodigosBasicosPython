```python
#Polimorfismo
"""
Crear una clase envase de líquido, que contenga los atributos: capacidad máxima y contenido actual
Debe contener el siguiente método:
- Rellenar: recibe un valor y debe aumentar dicha cantidad el contenido actual sin sobrepasar la capacidad máxima

Además, de éste debe heredar dos clases hijas: vaso y botella
La clase botella debe contener el siguiente métodos:
- Servir: recibe un valor y debe reducir dicha cantidad el contenido actual hasta llegar a 0
"""

class EnvaseLiquido:
    
    #Constructor
    def __init__(self, capacidad, contenido):
        self.capacidad=capacidad
        self.contenido=contenido
        
    #Función rellenar
    def Rellenar(self,cantidad):
        self.contenido+=cantidad
        if(self.contenido>self.capacidad):
            self.contenido=self.capacidad
        #print(f"La botella contiene ahora {self.contenido} cl\n")
        
    #Función __add__
    def __add__(self, other):
        if(type(self)==type(other)):
            return(EnvaseLiquido(self.capacidad+other.capacidad,self.contenido+other.contenido))
        else:
            print("Error: No se pueden sumar porque no son de la misma clase\n")
        
    #Función __str__
    def __str__(self):
        return(f"Envase de líquidos con una capacidad máxima de {self.capacidad} cl y un contenido actual de {self.contenido} cl\n")
        
class Vaso(EnvaseLiquido):
    
    #Constructor (contenido por defecto a 0)
    def __init__(self, capacidad, contenido=0):
        super().__init__(capacidad, contenido)
        #print("Vaso creado\n")
        
    #Función __add__
    def __add__(self, other):
        if(type(self)==type(other)):
            return(Vaso(self.capacidad+other.capacidad,self.contenido+other.contenido))
        else:
            print("Error: No se pueden sumar porque no son de la misma clase\n")
        
    #Función __str__
    def __str__(self):
        return(f"Vaso con una capacidad máxima de {self.capacidad} cl y un contenido actual de {self.contenido} cl\n")
        
class Botella(EnvaseLiquido):
    
    #Constructor (contenido por defecto a 0)
    def __init__(self, capacidad, contenido=0):
        super().__init__(capacidad, contenido)
        #print("Botella creada\n")
        
    #Función servir
    def Servir(self,cantidad):
        if(self.contenido<cantidad):
            #Variable para saber cuánto se ha servido
            servido=self.contenido
            self.contenido=0
        else:
            self.contenido-=cantidad
            #Variable para saber cuánto se ha servido
            servido=cantidad
        #print(f"Se han servido {servido} cl.\nLa botella contiene ahora {self.contenido} cl\n")
        
    #Función __add__
    def __add__(self, other):
        if(type(self)==type(other)):
            return(Botella(self.capacidad+other.capacidad,self.contenido+other.contenido))
        else:
            print("Error: No se pueden sumar porque no son de la misma clase\n")
        
    #Función __str__
    def __str__(self):
        return(f"Botella con una capacidad máxima de {self.capacidad} cl y un contenido actual de {self.contenido} cl\n")
        
#Función enjuagar       
def Enjuagar(el):
    el.contenido=0
    print("Envase vaciado")
    el.Rellenar(el.capacidad)
    print("Envase completamente lleno de agua")
    el.contenido=0
    print("Agua quitada y envase limpiado\n")
    
        
#Creación de un envase "genérico"
el1=EnvaseLiquido(50,25)
#Validación de la función __str__
print(el1)
#Validación de la función __add__
el2=EnvaseLiquido(20,5)
el3=el1+el2
print(el3)

print("\n----------\n")

#Creación de un vaso
v1=Vaso(50)
#Validación de la función __str__
print(v1)
#Validación de la función __add__
v2=Vaso(20,5)
v3=v1+v2
print(v3)

print("\n----------\n")

#Creación de una botella
b1=Botella(50)
#Validación de la función __str__
print(b1)
#Validación de la función __add__
b2=Botella(20,5)
b3=b1+b2
print(b3)

print("\n----------\n")

#Validación de la función __add__ con clases diferentes
el4=el3+b1
#Creación de un envase líquido a partir de objetos de otra clase mediante la función __add__
el5=EnvaseLiquido.__add__(b1,b3)
print(el5)

print("\n----------\n")

#Validación de la función enjuagar
for i in [el3,b3,v3]:
    print("Enjuagando nuevo envase")
    Enjuagar(i)
```
