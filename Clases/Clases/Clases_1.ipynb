```python
#Clases
"""
Crear una clase sujetapapeles, que contenga los atributos: peso, cantidad de papeles a sujetar y colocado.
Además, debe contener los siguientes métodos:
- Colocar: debe colocar el objeto si no lo está y dejarlo si ya lo está
- Quitar: debe quitar el objeto si está colocado y dejarlo si no es el caso
- Añadir papeles: debe añadir papeles al montón a sujetar
- Quitar papeles: debe extraer papeles al montón a sujetar y dejarlo en 0 si hay menos de los que se quieren quitar
- Comprobar: debe tomar un valor numérico (que representa la fuerza del viento) y, mediante, una fórmula que tenga en cuenta éste, la cantidad de
  de papeles y el peso, comprobar si es suficiente para sujetarlos en caso de que esté colocado
"""

class SujetaPapeles:
    
    #Constructor (colocado por defecto en True)
    def __init__(self, peso, papeles, colocado=True):
        self.peso=peso
        self.papeles=papeles
        self.colocado=colocado
        print("Portapapeles creado\n")
        
    #Función colocar
    def Colocar(self):
        self.colocado=True
        print("Portapapeles colocado\n")
                
    #Función quitar
    def Quitar(self):
        self.colocado=False
        print("Portapapeles quitado\n")
        
    #Función añadir papeles
    def MasPapeles(self, cantidad):
        self.papeles+=cantidad
        print(f"Cantidad de papeles: {self.papeles}\n")
        
    #Función quitar papeles
    def MenosPapeles(self, cantidad):
        self.papeles-=cantidad
        if(self.papeles<0):
            self.papeles=0
        print(f"Cantidad de papeles: {self.papeles}\n")
        
    #Función comprobar
    def Comprobar(self, viento):
        #Comprobación de que esté colocado
        if(self.colocado):
            #La fórmula que me he inventado es que el viento por la cantidad de papeles entre 10 debe ser mayor que el peso para que no aguente
            #Ésta sería la variable que contendría el valor resultante de la multiplicación y división
            fuerza=viento*self.papeles/10
            if(fuerza>self.peso):
                print(f"El portapapeles no ha aguantado\n{self.papeles} papeles han caído al suelo\n")
                self.papeles=0
            else:
                print("El portapapeles ha aguantado\n")
        else:
            #Comprobación de que hubiera algún papel que sujetar
            if(self.papeles>0):
                print(f"El portapapeles no estaba colocado\n{self.papeles} papeles han caído al suelo\n")
                self.papeles=0
            else:
                print(f"El portapapeles no estaba colocado pero no había nada que sujetar\n")
            
#Creación de un sujetapepeles con el valor por defecto
sp1=SujetaPapeles(20,5)

#Validación de la función quitar
sp1.Quitar()
print("#\n",sp1.colocado,"\n#")

#Validación de la función colocar
sp1.Colocar()
print("#\n",sp1.colocado,"\n#")

#Validación de la función quitar papeles
sp1.MenosPapeles(3)
print("#\n",sp1.papeles,"\n#")

#Validación de la función añadir papeles
sp1.MasPapeles(10)
print("#\n",sp1.papeles,"\n#")

#Validación de la función comprobar con viento insuficiente
sp1.Comprobar(5)

#Validación de la función comprobar con demasiado viento
sp1.Comprobar(50)
print("#\n",sp1.papeles,"\n#")

            
#Creación de un sujetapepeles sin usar el valor por defecto
sp2=SujetaPapeles(20,5,False)

#Validación de la función quitar papeles, con demasiada cantidad
sp2.MenosPapeles(40)
print("#\n",sp1.papeles,"\n#")


#Validación de la función comprobar estando sin colocar y sin nigún papel
sp2.Comprobar(5)

#Validación de la función comprobar estando sin colocar y con algún papel
sp2.MasPapeles(10)
sp2.Comprobar(5)
print("#\n",sp1.papeles,"\n#")
```
